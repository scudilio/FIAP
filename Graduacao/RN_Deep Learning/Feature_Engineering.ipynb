```python
#substituindo precos < 10 pela media
media_price = df['price'].mean()
```

```python
df['price'] = df['price'].apply(lambda x: media_price if x<10 else x)
```

```python
#precos >350000 excluir
df = df.loc[df.price<350000]
```
