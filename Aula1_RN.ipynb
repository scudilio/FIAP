## **Criando o while**

```python
import math

x = 0
y = 0

w0 = 0.5

def ativacao(soma):
  if soma>= 0 :
    return 1
  else:
    return 0

print ("O valor de entrada foi:", x)
print ("O valor real, desejado era:", y)

taxa_aprendizado = 0.01
erro = math.inf
interacao = 0 # qual foi a demora





while not erro == 0:
  interacao += 1
  print("Interação",interacao)
  print("Peso",w0)
  soma = (w0 * x)

  y_hat = ativacao(soma)

  print ("O resultado da predição foi:", y_hat)

  erro = y - y_hat
  print("O erro foi ", erro)


  if not erro == 0:
    w0 = w0 + (taxa_aprendizado * x * erro)
print ("A rede neural aprendeu")
```

## **ADICIONANDO O VIÉS**

```python
import math

x = 0
y = 0

w0 = 0.5

def ativacao(soma):
  if soma>= 0 :
    return 1
  else:
    return 0

print ("O valor de entrada foi:", x)
print ("O valor real, desejado era:", y)

taxa_aprendizado = 0.01
erro = math.inf
interacao = 0 # qual foi a demora

bias = 1
bias_weight = 0.5



while not erro == 0:
  interacao += 1
  print("Interação",interacao)
  print("Peso",w0)
  soma = (w0 * x)  + (bias * bias_weight)

  y_hat = ativacao(soma)

  print ("O resultado da predição foi:", y_hat)

  erro = y - y_hat
  print("O erro foi ", erro)


  if not erro == 0:
    w0 = w0 + (taxa_aprendizado * x * erro)
    bias_weight = bias_weight + (taxa_aprendizado * bias * erro)
print ("A rede neural aprendeu")
```

```python
import numpy as np

#Definição das variaveis da RNA

entradas = np.array([[0,0],[0,1],[1,0],[1,1],[1,1]])



saidas = np.array([0,0,0,1,1])



pesos=np.array([0.0,0.0])



taxa = 0.1



epocas = 5



#Função de ativação

def stepFunction(soma):

    if soma >= 1:

        return 1



    return 0





#Função para calcular o output

def calcularOutput(registros):

    s= registros.dot(pesos)

    return stepFunction(s)





#Função Para treinar o Array

def treinarPerceptron():



    count=0

    #Loop de incio de treino da RNA

    while (count <= epocas):



        #Zerando o erro para que sempre inicie o teste sem erros

        erroTotal = 0



        #Loop para o cálculo do erro em cada uma das saidas desejadas com as saidas calculadas

        for i in range(len(saidas)):

            saidaCalculada = calcularOutput(np.asarray(entradas[i]))

            erro = saidas[i] - saidaCalculada



            #Redefinindo o Valor do erro novamente

            erroTotal += erro

            if (erroTotal != 0):



                #Atualizando todos  os pesos caso exista erro

                for j in range(len(pesos)):



                    #Função de Ajuste de peso

                    pesos[j]= pesos[j]+(taxa * entradas[i][j] * erro)



                    print ("Pesos Atualizados :" + str(pesos[j]))

        print ("Total de erros: " + str(erroTotal))

        count += 1

#Chamada do treino

treinarPerceptron()
```

## **TREINANDO REDE PERCEPTRON COM MAIS DE UM NEURÔNIO**
